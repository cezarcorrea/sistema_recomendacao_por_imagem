```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import GlobalAveragePooling2D
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity # Ou euclidean_distances

# Define o caminho para o seu dataset no Google Drive
DATASET_PATH = '/content/drive/MyDrive/sistema_recomendacao/dataset'
IMG_HEIGHT, IMG_WIDTH = 224, 224
BATCH_SIZE = 32
EPOCHS = 10

print("Bibliotecas instaladas e Drive montado.")

# --- Célula 2: Carregamento e Pré-processamento do Modelo ---

# Carrega o modelo ResNet50 pré-treinado no ImageNet, sem a camada de classificação final
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))

# Adiciona uma camada de pooling global para obter um vetor de características de tamanho fixo
x = base_model.output
x = GlobalAveragePooling2D()(x)
model = Model(inputs=base_model.input, outputs=x)

print("Modelo ResNet50 carregado e configurado para extração de características.")
model.summary()

# --- Célula 3: Extração de Características das Imagens ---

def extract_features(img_path, model):
    """Extrai características de uma única imagem."""
    try:
        img = image.load_img(img_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) # Adiciona a dimensão do batch
        img_array = preprocess_input(img_array) # Normaliza os pixels
        features = model.predict(img_array, verbose=0)
        return features.flatten() # Retorna um vetor 1D
    except Exception as e:
        print(f"Erro ao processar imagem {img_path}: {e}")
        return None

# Lista de todos os arquivos de imagem no dataset
image_paths = []
labels = []
for class_name in os.listdir(DATASET_PATH):
    class_dir = os.path.join(DATASET_PATH, class_name)
    if os.path.isdir(class_dir):
        for img_name in os.listdir(class_dir):
            if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_paths.append(os.path.join(class_dir, img_name))
                labels.append(class_name)

print(f"Encontradas {len(image_paths)} imagens em {len(set(labels))} classes.")

# Extrai características para todas as imagens
image_features = []
valid_image_paths = [] # Para guardar os caminhos das imagens que foram processadas com sucesso
for img_path in image_paths:
    features = extract_features(img_path, model)
    if features is not None:
        image_features.append(features)
        valid_image_paths.append(img_path)

image_features = np.array(image_features)
labels = np.array([labels[i] for i in range(len(image_paths)) if image_features[i] is not None]) # Filtra labels de acordo com features válidas

print(f"Características extraídas para {len(image_features)} imagens.")

# --- Célula 4: Cálculo de Similaridade e Geração de Recomendações ---

def get_recommendations(query_img_path, top_n=5):
    """
    Gera recomendações de imagens similares a uma imagem de consulta.
    """
    query_features = extract_features(query_img_path, model)
    if query_features is None:
        print("Não foi possível extrair características da imagem de consulta.")
        return []

    # Calcula a similaridade de cosseno entre a imagem de consulta e todas as outras imagens
    # cosine_similarity espera arrays 2D, por isso usamos [query_features] e [image_features]
    similarities = cosine_similarity([query_features], image_features)[0]

    # Obtém os índices das imagens mais similares (ignorando a própria imagem de consulta, se ela estiver no dataset)
    # Criamos um array de tuplas (similaridade, índice) e ordenamos
    sim_indices = sorted(enumerate(similarities), key=lambda x: x[1], reverse=True)

    recommendations = []
    for i, (idx, sim) in enumerate(sim_indices):
        if i >= top_n + 1: # Pegamos top_n + 1 para poder pular a própria imagem de consulta
            break
        # Verifica se o caminho da imagem de consulta é diferente do caminho da imagem no dataset
        if valid_image_paths[idx] != query_img_path:
            recommendations.append({
                'image_path': valid_image_paths[idx],
                'similarity': sim,
                'label': labels[idx]
            })
            if len(recommendations) >= top_n:
                break

    return recommendations

# --- Célula 5: Exemplo de Uso e Visualização ---

# Este caminho aponta para uma imagem no seu dataset que você quer usar como consulta.
QUERY_IMAGE_PATH = '/content/drive/MyDrive/sistema_recomendacao/dataset/relogios/relogio (7).png'

if os.path.exists(QUERY_IMAGE_PATH):
    print(f"Gerando recomendações para: {QUERY_IMAGE_PATH}")
    recommended_items = get_recommendations(QUERY_IMAGE_PATH, top_n=5)

    # Visualiza a imagem de consulta e as recomendações
    plt.figure(figsize=(15, 8))

    # Imagem de consulta
    plt.subplot(2, 3, 1)
    img = image.load_img(QUERY_IMAGE_PATH, target_size=(IMG_HEIGHT, IMG_WIDTH))
    plt.imshow(img)
    plt.title(f"Buscado: {os.path.basename(QUERY_IMAGE_PATH)}")
    plt.axis('off')

    # Recomendações
    for i, rec in enumerate(recommended_items):
        plt.subplot(2, 3, i + 2)
        img_rec = image.load_img(rec['image_path'], target_size=(IMG_HEIGHT, IMG_WIDTH))
        plt.imshow(img_rec)
        plt.title(f"{rec['label']} (Sim: {rec['similarity']:.2f})")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

    print("\nRecomendações:")
    for rec in recommended_items:
        print(f"- Imagem: {os.path.basename(rec['image_path'])}, Classe: {rec['label']}, Similaridade: {rec['similarity']:.4f}")
else:
    print(f"Erro: A imagem de consulta '{QUERY_IMAGE_PATH}' não foi encontrada. Verifique o caminho e tente novamente.")
```
